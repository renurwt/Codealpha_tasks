```python
# Exploratory Data Analysis (EDA) on COVID Dataset

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
df = pd.read_csv("/owid-covid-data.csv.zip")
print("First 5 Rows:")
print(df.head())
print("\n Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nSummary Statistics:")
print(df.describe(include='all'))

print("\n  Missing Values:")
print(df.isnull().sum())
```

```python
# Records Count by Continent
plt.figure(figsize=(6,4))
sns.countplot(x='continent', data=df, palette='Set2')
plt.title('Record Count by Continent')
plt.xlabel('Continent')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of New Cases
plt.figure(figsize=(7,5))
sns.histplot(df['new_cases'].dropna(), bins=50, kde=True, color='skyblue')
plt.title('Distribution of New Cases (log scale)')
plt.xlabel('New Cases per Day')
plt.ylabel('Frequency')
plt.xscale("log")
plt.show()
```

```python
# New Cases by Continent
plt.figure(figsize=(7,5))
sns.boxplot(x='continent', y='new_cases', data=df, palette='pastel')
plt.ylim(0, df['new_cases'].quantile(0.95))   # cut extreme spikes
plt.title('New Cases Distribution by Continent')
plt.xlabel('Continent')
plt.ylabel('New Cases')
plt.show()
```

```python
if "people_fully_vaccinated_per_hundred" in df.columns:
    plt.figure(figsize=(7,5))
    sns.barplot(
        x='continent',
        y='people_fully_vaccinated_per_hundred',
        data=df,
        estimator=lambda x: x.mean(),
        ci=None,
        palette='muted'
    )
    plt.title('Average % Fully Vaccinated by Continent')
    plt.xlabel('Continent')
    plt.ylabel('% Fully Vaccinated')
    plt.show()
```
